```python
from torch import nn
from torchsummary import summary

class DwConv(nn.Module):
    def __init__(self, inputs, outputs):
        super().__init__()
        norm_layer = nn.BatchNorm2d
        self.depthwise = nn.Conv2d(inputs, inputs, kernel_size=3, padding='same', groups=inputs)
        self.relu = nn.ReLU()
        self.bn1 = norm_layer(inputs)
        self.pointwise = nn.Conv2d(inputs, outputs, kernel_size=1)
        self.bn2 = norm_layer(outputs)

    def forward(self, x):
        out = self.depthwise(x)
        out = self.bn1(out)
        out = self.relu(out)
        out = self.pointwise(out)
        out = self.bn2(out)
        out = self.relu(out)
        return out

# Проверка  
block = DwConv(3, 64)
summary(block, input_size=(3, 224, 224), device = 'cpu')
```
